```python
import numpy as np
plan = np.genfromtxt("/content/drive/MyDrive/Housing Price data set.csv", delimiter=",", skip_header=1)
```

```python


cost = []
for i in range(len(plan)):
	cost.append(plan[i][1])
 
t_cost=[]
for i in range(383,546):
    t_cost.append(plan[i][1])

plotarea = []
for i in range(len(plan)):
	plotarea.append(plan[i][2])
 
t_plotarea=[]
for i in range(383,546):
    t_plotarea.append(plan[i][2])

bedroom_no = [] 
for i in range(len(plan)):
	bedroom_no.append(plan[i][3])
 
t_bedroom_no=[]
for i in range(383,546):
    t_bedroom_no.append(plan[i][3])

bathroom_no = []
for i in range(len(plan)):
	bathroom_no.append(plan[i][4])

t_bathroom_no=[]
for i in range(383,546):
    t_bathroom_no.append(plan[i][4])
```

```python
c1 = []

for i in range(len(cost)):
    c1.append(1)
c1

zip_concat = zip(c1,plotarea,bedroom_no,bathroom_no)
X = list(zip_concat)
X_T = np.transpose(X)
X_X_T = np.dot(X_T,X)
X_X_T_1= np.linalg.inv(X_X_T)
X_X_T_X_T_1 = np.dot(X_X_T_1,X_T)
Y = cost
_Y = np.dot(X_X_T_X_T_1,Y)
```

```python
print("Without Normalisation : Prediction of Data : ",end=" \n")
def without(X1,X2,X3):
    res = _Y[0]+X1*_Y[1]+X2*_Y[2]+X3*_Y[3]
    return res
Ypredict1 = []
for i in range(0,163):
    value = without(t_plotarea[i],t_bedroom_no[i],t_bathroom_no[i])
    Ypredict1.append(value)
Ypredict1
```

```python
error1 =0 
for i in range(163):
    error1 = error1 + abs((Ypredict1[i] - t_cost[i]) / t_cost[i])
error1 = error1 / 163
print("Error In Prediction without Normalisation",error1 * 100)
```

```python
print("With Normalisation : Prediction of Data : " , end=" \n")
id = np.identity(4, dtype=float)
id[0][0] = 0

for i in range(30):
    b_X_X_T = np.add(X_X_T,id * i)
    b_X_X_T_1= np.linalg.inv(b_X_X_T)
    temp = np.dot(X_T,Y)
    _Y2 = np.dot(b_X_X_T_1,temp)
    Ypredict2 = []
```

```python
def withn(X1,X2,X3):
    res = _Y2[0]+X1*_Y2[1]+X2*_Y2[2]+X3*_Y2[3]
    return res
```

```python
for i in range(163):
    value = withn(t_plotarea[i],t_bedroom_no[i],t_bathroom_no[i])
    Ypredict2.append(value)
Ypredict2
```

```python
error2 = 0
for i in range(163):
    error2 = error2 + abs((Ypredict2[i] - t_cost[i]) / t_cost[i])
error2 = error2 /163
   
print("Error In Prediction with Normalisation",error2 * 100)
```
